```python
import networkx as nx
```

```python
Gragh = nx.Graph()
Gragh.add_nodes_from(['a', 'b', 'c', 'd'])
Gragh.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('a','c')])
nx.draw(Gragh, with_labels = True)
```

```python
list(Gragh.neighbors('b'))
```

```python
for neighbor in Gragh.neighbors('b'):
    print(neighbor)
```

```python
Gragh.has_node('a')
```

```python
nx.is_tree(Gragh)
```

```python
nx.is_connected(Gragh)
```

```python
Gragh.degree('a')
```

```python
DGragh = nx.DiGraph()
DGragh.add_edges_from([('1','2'), ('2','3'), ('1','4'), ('3','4')])
nx.draw(DGragh, with_labels = True)
```

```python
G2=nx.Graph()
G2.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(G2,with_labels=True,node_size=500)
```

```python
dict(G2.degree())
```

```python
nx.to_numpy_matrix(G2)
```
